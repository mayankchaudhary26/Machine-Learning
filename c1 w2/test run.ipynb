```python
import turicreate
```

```python
sale = turicreate.SFrame('home_data.sframe')
```

```python
sale
```

```python
turicreate.show(sale[1:50]['sqft_living'],sale[1:50]['price'])
```

```python
training_set, test_set = sale.random_split(.8,seed=0)
```

```python
sqft_model = turicreate.linear_regression.create(training_set,target='price',features=['sqft_living'])
```

```python
my_features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','zipcode']
```

```python
sale[my_features].show()
```

```python
turicreate.show(sale['zipcode'],sale['price'])
```

```python
my_features_model = turicreate.linear_regression.create(training_set,target='price',features=my_features)
```

```python
print (my_features)
```

```python
print (sqft_model.evaluate(test_set))
print (my_features_model.evaluate(test_set))
```
